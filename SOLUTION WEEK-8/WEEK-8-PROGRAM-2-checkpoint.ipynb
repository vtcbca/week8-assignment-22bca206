### C. Write a python scripts that read the sales.csv file and plot a bar chart that shows totalsales of the year. 

```python
import pandas 
```

```python
file_df=pandas.read_csv('C:\sqlite3\sales.csv')
file_df
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.figure(figsize=(10, 6))
plt.bar(file_df['year'],file_df['totalsales'],color='r', alpha=0.7)
plt.title('Total Sales For Year',fontsize=25,pad=20)
plt.xlabel('----- Year -----',fontsize=14,labelpad=20)
plt.ylabel('----- Total Sale -----',fontsize=14,labelpad=20)
plt.show()
```
