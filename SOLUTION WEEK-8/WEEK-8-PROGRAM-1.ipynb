```python
import csv
import pandas
```

```python
heading=['Prod_Name','Jan','Feb','Mar','Apr','May','Jun']

def input_data_from_user():
    list1=[]
    for i in range(int(input("How Many Records You Want To Insert into File:"))):
        pname=input('Enter Product Name : ')
        jan=int(input("How Many Product Sold In January ?: "))
        feb=int(input("How Many Product Sold In Fabruary ?: "))
        mar=int(input("How Many Product Sold In March ?: "))
        apr=int(input("How Many Product Sold In April ?: "))
        may=int(input("How Many Product Sold In may ?: "))
        june=int(input("How Many Product Sold In June ?: "))
        l=[pname,jan,feb,mar,apr,may,june]
        list1.append(l) 
    return list1

with open(r'D:\Product_Details_of_chocolates.csv','w',newline='') as file:
    obj=csv.writer(file)
    obj.writerow(heading)
    obj.writerows(input_data_from_user())
```

```python
df=pandas.read_csv(r'D:\Product_Details_of_chocolates.csv')
df
```

```python
total=df['Jan']+df['Feb']+df['Mar']+df['Apr']+df['May']+df['Jun']
df['Total_Sell']=total
avg=total/6
df['Average_Sell']=avg
df
```

### C. Plot Total sell and average sell together on line chart with proper Legends, titles and lables. 

```python
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
seaborn.set_style('whitegrid')
plt.plot(df['Prod_Name'],df['Total_Sell'],marker='o')
plt.plot(df['Prod_Name'],df['Average_Sell'],marker='o')
plt.xticks(rotation=60)
plt.xlabel('Product Name')
plt.ylabel('Sell Of Product')
plt.title('Total And Average Sell Of Product')
plt.legend(['Total_Sell','Average_Sell'])
plt.subplots_adjust(bottom=0.25)
plt.show()
```
